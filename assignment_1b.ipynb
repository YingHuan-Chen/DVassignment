```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from scipy import linalg
%matplotlib inline

def show_gradient_descent(A, b):
#  enter your code here, and remove the following line (30%)
  x_lists=[]
  y_lists=[]
  f_lists=[]
  x=np.array([[-2],[2]])
  alpha=0.02
  max_iter=1000
  f=(np.linalg.norm(np.dot(A,x)-b))**2
  x_lists.append(x[0][0])
  y_lists.append(x[1][0])
  f_lists.append(f)
  for k in range(max_iter):
    print(k,'(','%.4f'%x[0][0],',','%.4f'%x[1][0],')','%.8f'%f)
    x_prev = x
    f_gradient = (np.dot(np.dot(A.T,A),x)-np.dot(A.T,b))*2
    x = x-alpha*f_gradient
    if np.linalg.norm(x-x_prev)<=1e-8:
      break
    f=(np.linalg.norm(np.dot(A,x)-b))**2
    x_lists.append(x[0][0])
    y_lists.append(x[1][0])
    f_lists.append(f)
  return (x_lists,y_lists,f_lists)
    
```

```python
def draw_gradient_descent(A, b, x_lists, y_lists, f_lists):
  xx, yy = np.meshgrid(np.arange(-2,2.1,0.1), np.arange(-2,2.1,0.1))
  F1 = A[0][0]*xx + A[0][1]*yy - b[0];
  F2 = A[1][0]*xx + A[1][1]*yy - b[1];
  zz = F1*F1 + F2*F2;
  plt3d = plt.figure().gca(projection='3d')
  plt3d.plot_surface(xx, yy, zz, alpha=0.7, cmap=cm.coolwarm)
    
#  enter your code here (14%)
  plt3d.scatter(x_lists,y_lists,f_lists,c='red')
  plt3d.plot(x_lists,y_lists,0,c='blue')

  plt.show()
```

```python
if __name__ == '__main__':
  b = np.array([[1], [2]])
  A = np.array([[1, -2], [1, -2]])
  lists=show_gradient_descent(A, b)
  x_lists=lists[0]
  y_lists=lists[1]
  f_lists=lists[2]
  #print(x_lists)
  draw_gradient_descent(A, b,x_lists,y_lists,f_lists)
```
